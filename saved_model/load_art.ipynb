```python
import wandb

# Set your W&B entity and project name
ENTITY = "hanchen-david-wang-vanderbilt-university"
PROJECT = "EXACT"
TYPE = 'model'
api = wandb.Api()
collections = [
    coll for coll in api.artifact_type(type_name=TYPE, project=PROJECT).collections()
]

```

```python
aliases = set()
for coll in collections:
    aliases.update(coll.aliases)
    print(coll, coll.aliases)
    if len(coll.aliases) == 0:
        aliases.add('')
    

print(collections)
print(aliases)
```

```python
import wandb
import os
# Define download directory
DOWNLOAD_PATH = "./wandb_artifacts"
os.makedirs(DOWNLOAD_PATH, exist_ok=True)

# Initialize W&B API
api = wandb.Api()
alias = 'latest'
# Download all artifacts
for coll in collections:
    for alias in coll.aliases:
        try:
            # Fetch the artifact using its alias
            artifact = api.artifact(f"{ENTITY}/{PROJECT}/{coll.name}:{alias}")

            if 'subject' not in coll.name:
                folder_name = '_'.join(coll.name.split('-')[0:2])
                artifact_dir = artifact.download(root=os.path.join(DOWNLOAD_PATH, folder_name))
            else:
                folder_name = '_'.join(coll.name.split('-')[0:2])
                # Download the artifact
                artifact_dir = artifact.download(root=os.path.join( DOWNLOAD_PATH, 'subjects', folder_name))

            # Move all .pth files to a single directory

        except Exception as e:
            print(f"Error downloading {coll.name}:{alias} - {e}")

print("All .pth files downloaded successfully!")
```
