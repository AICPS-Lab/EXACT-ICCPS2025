```python
import os
import numpy as np
import pandas as pd
import pickle

# Specify the columns to select
columns_to_select = ['AccX','AccY','AccZ','GyroX','GyroY','GyroZ']
def get_digits(str1):
    c = ""
    for i in str1:
        if i.isdigit():
            c += i
    return c
def sort_key(filename):
    parts = filename.rstrip('.csv').split('_')
    # Convert parts to integers where applicable for proper numeric sorting
    res = []
    for part in parts: 
        # print('s', get_digits(part))
        r = get_digits(part)
        if r == "":
            if part == 'L':
                res.append(0)
            elif part == 'R':
                res.append(1)
            else:
                AssertionError()
            # res.append(part)
        else:
            res.append(int(r))
    return res
def process_files(root_dir):
    data_dict = {}
    data_dict['X'] = []
    data_dict['subject'] = []
    data_dict['exercise'] = []
    data_dict['repetition'] = []
    for dirpath, dirnames, filenames in os.walk(root_dir):
        all_files = sorted(filenames, key=sort_key)
        print(all_files)
        for filename in all_files:
            # Skip files that are not CSV
            if not filename.endswith(".csv"):
                continue

            # Construct full file path
            filepath = os.path.join(dirpath, filename)

            # Load CSV file and select required columns
            df = pd.read_csv(filepath)
            df_selected = df[columns_to_select].values

            # Use filename as key (without extension)
            keys = filename.split('.')[0].split('_')
            data_dict['X'].append(df_selected)
            if keys[0].isdigit():
                keys[0] = int(keys[0])
            if keys[1].isdigit():
                keys[1] = int(keys[1])
            if keys[3].isdigit():
                keys[3] = int(keys[3])
            data_dict['subject'].append(keys[0])
            data_dict['exercise'].append(keys[1])
            data_dict['repetition'].append(keys[3])

    return data_dict

data_dict = process_files('./PHYSIQ_segmented')

# Save dictionary with pickle
with open('PHYSIQ.pickle', 'wb') as f:
    pickle.dump(data_dict, f)
```
